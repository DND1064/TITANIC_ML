```python
titanic_data['Embarked'].fillna(method='ffill', inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(titanic_data, titanic_data["Survived"]):
    strat_train_set = titanic_data.loc[train_index].copy()
    strat_test_set = titanic_data.loc[test_index].copy()
```

```python
plt.subplot(1, 2, 1)
strat_train_set['Survived'].hist()
strat_train_set['Pclass'].hist()

plt.subplot(1, 2, 2)
strat_test_set['Survived'].hist()
strat_test_set['Pclass'].hist()
plt.show()
```

```python
print(strat_train_set.isnull().sum())
print(strat_test_set.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

age_imputer = SimpleImputer(strategy='mean')
strat_train_set['Age'] = age_imputer.fit_transform(strat_train_set[['Age']])
strat_test_set['Age'] = age_imputer.transform(strat_test_set[['Age']])
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
strat_train_set['Sex'] = le_sex.fit_transform(strat_train_set['Sex'])
strat_test_set['Sex'] = le_sex.transform(strat_test_set['Sex'])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
embarked_train_encoded = ohe.fit_transform(strat_train_set[['Embarked']])
embarked_test_encoded = ohe.transform(strat_test_set[['Embarked']])
embarked_cols = ohe.get_feature_names_out(['Embarked'])
embarked_df_train = pd.DataFrame(embarked_train_encoded, columns=embarked_cols, index=strat_train_set.index)
embarked_df_test = pd.DataFrame(embarked_test_encoded, columns=embarked_cols, index=strat_test_set.index)
strat_train_set = pd.concat([strat_train_set, embarked_df_train], axis=1)
strat_test_set = pd.concat([strat_test_set, embarked_df_test], axis=1)
```

```python
# Drop unwanted columns from the dataset
drop_cols = ["Embarked", "Name", "Ticket", "Cabin", "Fare"]
strat_train_set.drop(columns=drop_cols, inplace=True, errors="ignore")
strat_test_set.drop(columns=drop_cols, inplace=True, errors="ignore")
```

```python
strat_train_set, strat_test_set = strat_train_set.align(strat_test_set, join='left', axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler
X_train = strat_train_set.drop("Survived", axis=1)
y_train = strat_train_set["Survived"]
X_test = strat_test_set.drop("Survived", axis=1)
y_test = strat_test_set["Survived"]

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [10, 100, 200, 500],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 3, 4]
}
clf = RandomForestClassifier()
grid_search = GridSearchCV(clf, param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)
```

```python
final_model = grid_search.best_estimator_
print("Test accuracy:", final_model.score(X_test_scaled, y_test))
```

```python
# === Final training on full data ===
# Clean full training data
titanic_data['Age'] = age_imputer.fit_transform(titanic_data[['Age']])
titanic_data['Sex'] = le_sex.transform(titanic_data['Sex'])
titanic_data['Embarked'].fillna(method='ffill', inplace=True)

embarked_encoded_full = ohe.fit_transform(titanic_data[['Embarked']])
embarked_df_full = pd.DataFrame(embarked_encoded_full, columns=ohe.get_feature_names_out(['Embarked']), index=titanic_data.index)
titanic_data = pd.concat([titanic_data, embarked_df_full], axis=1)

titanic_data.drop(columns=drop_cols, inplace=True, errors="ignore")

X_final = titanic_data.drop("Survived", axis=1)
y_final = titanic_data["Survived"]
```

```python
X_final_scaled = scaler.fit_transform(X_final)
grid_search.fit(X_final_scaled, y_final)
prod_final_model = grid_search.best_estimator_
```

```python
# === Process test data ===
titanic_test_data = pd.read_csv("data/test.csv")
titanic_test_data['Embarked'].fillna(method='ffill', inplace=True)
titanic_test_data['Age'] = age_imputer.transform(titanic_test_data[['Age']])
titanic_test_data['Sex'] = le_sex.transform(titanic_test_data['Sex'])

embarked_test_encoded = ohe.transform(titanic_test_data[['Embarked']])
embarked_df_test_final = pd.DataFrame(embarked_test_encoded, columns=ohe.get_feature_names_out(['Embarked']), index=titanic_test_data.index)
titanic_test_data = pd.concat([titanic_test_data, embarked_df_test_final], axis=1)

titanic_test_data.drop(columns=drop_cols, inplace=True, errors="ignore")
```

```python
# Align with training features
X_test_final = titanic_test_data.copy()
X_test_final, _ = X_test_final.align(X_final, join='right', axis=1, fill_value=0)

X_test_final_scaled = scaler.transform(X_test_final)

# Predict and save
predictions = prod_final_model.predict(X_test_final_scaled)
submission = pd.DataFrame({
    "PassengerId": titanic_test_data["PassengerId"],
    "Survived": predictions
})
submission.to_csv("data/predictions1.csv", index=False)

```
